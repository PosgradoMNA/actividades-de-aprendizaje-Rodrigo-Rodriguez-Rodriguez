7.Reduce las dimensiones con PCA, si consideras necesario.
Indica la varianza de los datos explicada por cada componente seleccionado. Para actividades de exploración de los datos la varianza > 70%
Indica la importancia de las variables en cada componente

```python
df1 = df.iloc[:,1:24]
df1
```

```python
print(df1)
```

```python
import numpy as np
from sklearn.decomposition import PCA
```

```python
Cards_df=df1
```

```python
pcs = PCA()
pcs.fit(Cards_df.iloc[:, 0:].dropna(axis=0))
pcsSummary_df = pd.DataFrame({'Standard deviation': np.sqrt(pcs.explained_variance_),
                             'Proportion of variance': pcs.explained_variance_ratio_,
                             'Cumulative proportion': np.cumsum(pcs.explained_variance_ratio_)
                             })
pcsSummary_df = pcsSummary_df.transpose()
pcsSummary_df
```

Puede verse en las 3 categorias que los primeros 9 atributos son las que más variabilidad tienen en nuestro dataset, lo que nos indica que contienen la mayor cantidad de información sin duplicidad.

8.Elabora los histogramas de los atributos para visualizar su distribución

```python
PC_components = np.arange(pcs.n_components_) + 1
#PC_components

_ = sns.set(style = 'whitegrid',
            font_scale = 1.2
            )

fig, ax = plt.subplots(figsize=(10, 7))

_ = sns.barplot(x = PC_components,
                y = pcs.explained_variance_ratio_,
                color = 'b'
                )

_ = sns.lineplot(x = PC_components-1,
                 y = np.cumsum(pcs.explained_variance_ratio_),
                 color = 'black',
                 linestyle = '-',
                 linewidth = 2,
                 marker = 'o',
                 markersize = 8
                 )

plt.title('Scree Plot')
plt.xlabel('N-th Principal Component')
plt.ylabel('Variance Explained')
plt.ylim(0, 1)
plt.show()
```

```python
pcsComponents_df = pd.DataFrame(pcs.components_.transpose(),
                                columns=pcsSummary_df.columns,
                                index=Cards_df.iloc[:, 0:].columns
                                )
pcsComponents_df.iloc[:,:8]
```
